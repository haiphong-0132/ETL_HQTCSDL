Kiểm tra danh mục Điện Tử - Điện Lạnh

```python
df8.loc[df8['Danh mục'] == 'Điện Tử - Điện Lạnh']
```

Từ danh mục ĐT-ĐL, ta có thể phân phối chúng vào 2 danh mục: Tivi thường (LED) và Smart Tivi - Android Tivi

```python
# Tivi thường (LED)
df8.loc[(df8['Danh mục'] == 'Điện Tử - Điện Lạnh') & (df8['Tên sản phẩm'].str.lower().str.contains('led')) & (~df8['Tên sản phẩm'].str.lower().str.contains('oled'))]
```

```python
# Smart Tivi - Android Tivi
df8.loc[(df8['Danh mục'] == 'Điện Tử - Điện Lạnh') & (df8['Tên sản phẩm'].str.lower().str.contains('smart') | df8['Tên sản phẩm'].str.lower().str.contains('android'))]
```

Kiểm tra danh mục Điện Tử - Điện Lạnh và Root

```python
df8.loc[df8['Danh mục'] == 'Root']
```

Từ danh mục Root, ta có thể phân phối chúng vào 2 danh mục: Tivi thường (LED) và Smart Tivi - Android Tivi

```python
df8.loc[(df8['Danh mục'] == 'Root') & (df8['Tên sản phẩm'].str.lower().str.contains('led')) & (~df8['Tên sản phẩm'].str.lower().str.contains('oled'))]
```

```python
df8.loc[(df8['Danh mục'] == 'Root') & (df8['Tên sản phẩm'].str.lower().str.contains('smart') | df8['Tên sản phẩm'].str.lower().str.contains('android'))]
```

$\rightarrow$ Nhìn chung vẫn giữ nguyên các danh mục: <br/>

**Smart Tivi - Android Tivi** (Bổ sung thêm Điện Tử - Điện Lạnh + Root)<br/>
**Tivi 4K**<br/> 
**Tivi OLED**<br/>
**Tivi thường (LED)** (Bổ sung thêm Điện Tử - Điện Lạnh + Root)<br/>
**Tivi QLED**<br/>
**Tivi cao cấp**<br/>
**Internet Tivi**<br/>
